```python
# ADJUST HEATMAP ZOOM, INTENSITY, AND POINT RADIUS

# Adusting heat_layer = gmaps,heatmap_layer() for dissapation- default of which is TRUE, set to false. add MAX_INTENSITY =300
# so that each measurement have a better gradient variance.  Finally, add POINT_RADIUS=4 to make each measurement radius larger.
# Doing that allows us to see a gradient of temperature and the spots *DONT* dissipate when you zoom in- they're fixed.

# Get the latitude and longitude.
locations = city_data_df[["Lat", "Lng"]]
# Get the maximum temperature.
max_temp = city_data_df["Max Temp"]
# Assign the figure variable.
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
# Assign the heatmap variable.
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp, dissipating = False, max_intensity = 300, point_radius = 4)
# Add the heatmap layer.
fig.add_layer(heat_layer)
# Call the figure to plot the data.
fig
```

```python
# Get travel destinations
# not adding hotel info to the *preferred_cities_df b/c thats the filtered DF, and needs to be filtered for each customer.
# Make a copy of preferred_cities_df and name is hotel_df.  ADD new column to the hotel_df to hold the name of the hotel.
# keeping columns - city, country, max temp, lat, and lng

hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# RETRIEVE HOTELS FROM A **NEARBY SEARCH**
# Using the lat and lng and specific parameters, use the Google places Nearby Search request to retrieve a hotel
# and add it to the column.

# First step is to set paramters for the search.
    # Go to Nearby Search Request Page
    # Nearby Search lets you perform these tasks.
# 1. Search for places within a specified area.
# 2. Refine teh search using KEYWORDS and specifying the TYPE of place you're looking for.
# 3. Use an API URL, where the output can be either XML or JSON

# SPECIFIC PARAMETERS YOU MUST APPLY - using the first 4 of those.
    # 1. API key.
    # 2. A *location* which will be the latitude and longitude
    # 3. A *radius* for the search. we're keeping ours to 5,000 meters.
    # 4. Type of Place (lodging)

# we'll need to look at the documentation on the Python Requests Library (Links to an external site.) provided earlier. 
# The section "Passing Parameters In URLs" states that we can add the parameters as a dictionary of strings,
# using the **params** keyword argument  **** requests.get('base URL', params=parameters) ****. 
```

```python
from config import g_key
```

```python
# Add the paramaters as key-value pairs. Add the *params* dictionary, API key, *type*, and *radius* parameters.

# Set parameters to search for a hotel.
params = {
    "radius" : 5000,
    "type" : "lodging",
    "key" : g_key}
```

```python
# PRACTICE IS IN GOOGLE_NEARBY_SEARCH NOTEBOOK ###############
```

```python
# Use base URL to search for hotels in Paris.
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
# Make request and get the JSON data from the search.
hotels = requests.get(base_url, params=params).json()
hotels
```

```python
# Iterate through hotel_df Dataframe
    # Using **iterrows() function to perform iteration, need to provide the *index* and the *row* in the **for** loop
    # with this syntax: for index, row in df.iterrows()

# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
  
```

```python
# Iterate through hotel_df Dataframe
    # Using **iterrows() function to perform iteration, need to provide the *index* and the *row* in the **for** loop
    # with this syntax: for index, row in df.iterrows()

# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found...skipping.")
```

```python
# now we want to add POP-UP MARKERS for each city that display the hotel name, city name, country, and maximum temperature.
# we need to add an **info_box_template** to do this.

# Syntax for info_box_template

# info_box_template = """
# <dl>
# <dt>Name</dt><dd>{column1}</dd>
# <dt>Another name</dt><dd>{column2}</dd>
# </dl>
# """
```

```python
# What the above (HTML) code means #### # What the above (HTML) code means ##### What the above (HTML) code means ####

# the info_box_template *variable* is assigned to a multiline string using three quotes ("""). 
# the text inside the multiline string is HTML code.
# HTML code is defined by the opening and closing of the < > brackets. e.g. - <tag> and <tag/>
# **Angular brackets always come in pairs.** The opening angular bracket is followed
# by some text inside, such as *dl*, *dt*, and *dd*. 
# the closing angular bracket is preceded by a foward-slash ("/")

# WHAT THOSE TAGS MEAN
# <dl> tag is a description list (dl)

# <dt> tag is a term or name in a descripton list that is *nested* under the <dl> tag

# <dd> tag is used to define the term or name or <dt> tag

# If we were to write out these tags on paper, it would look like this. (spacing lines up there for a reason)

# Description List: <dl>
#       Description Term: <dt>
#              Description Definition: <dd>

#       Description Term: <dt>
#              Description Definition: <dd>
```

```python
# For our purpose, we'll add the hotel name, city name, country code, and the maximum temperature values 
# from the *hotel_df* DataFrame as the description defintion.

info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

# Next, add the data to the code by iterating through the hotel_df DataFrame
# using the iterrows() function. Then add the information to the gmaps.marker_layer() attribute with the locations.

# According to the documentation (Links to an external site.), 
# we can use this example to create the data for the **info_box_template**:

# From the documentation, we can take this code:
# plant_info = [info_box_template.format(**plant) for plant in nuclear_power_plants]

# Then edit this code by replacing "plant_info" with "hotel_info," and then use the iterrows() function to get the index and data in the row to add to the marker.


# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
### WHAT IS GOING ON IN THE CODE ABOVE (COPIED BELOW AS WELL)

# Store the DataFrame Row.
# hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]

# 1. We set the **hotel_info** equal to the info_box_template. (content)

# 2. Using the list comprehension, **info_box_template.format(**row) for index, row in hotel_df.interrows()
# we iterate through each "row" of the **hotel_df** DataFrame and then **format** the **info_box_template**
# with the data we set to populate from each row.  ****REMEMBER*** not using *every row*- only using
# the rows defined in the **info_box_template** which are *Hotel Name*, *City*, *Country*, and *Max Temp*
```

```python
# Next, in the code we used to create the heatmap w/ markers, add **info_box_content = hotel_info** 
# to *gmaps.marker_layer()* attribute with the locations.

locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]
fig = gmaps.figure(center = (30.0, 31.0), zoom_level = 1.5)
heat_layer = gmaps.heatmap_layer(locations, weights = max_temp, dissipating = False,
                                max_intensity = 300, point_radius = 4)
marker_layer = gmaps.marker_layer(locations, info_box_content = hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data
fig
```
